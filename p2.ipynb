```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10

x = np.linspace(-10, 10, 400)
y = f(x)
plt.plot(x, y, label='función')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import bisect, newton

def f(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10

x1 = bisect(f, -10, -1) 
x2 = bisect(f, 0, 3)  
x3 = bisect(f, 4, 6)   
x4 = newton(f, 2,tol=0.01) 
x5 = newton(f, 1,tol=0.01)  

print("Raíces:")
print(f"x1 = {x1}")
print(f"x2 = {x2}")
print(f"x3 = {x3}")
print(f"x4 = {x4}")
print(f"x5 = {x5}")
```
